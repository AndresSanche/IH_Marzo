```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ironhack-labs/lab-random-variable-distributions/master/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
data
```

```python
data.info()
```

```python
numericals = data.select_dtypes('object')
numericals.columns
```

```python
for col in numericals.columns:
    if col != "Customer":
        sns.displot(numericals[col], kde=True)
        plt.title(col)
        plt.show()
```
