```python
def clean_gender(x):
    x = str(x)
    if x == "F":
        return 'F'
    elif x == 'M':
        return 'M'
    else: return 'Other'
```

```python
categorical['GENDER'] = list(map(clean_gender,categorical['GENDER']))
```
